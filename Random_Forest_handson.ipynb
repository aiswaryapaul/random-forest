```python
# selecting only the columns that we need to label encode

col_list = []

for i in df.columns:
  if ((df[i].dtype == 'object') & (i != 'Churn')):
      col_list.append(i)

col_list
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in col_list:
  df[i] = le.fit_transform(df[i])
```

```python
df.head(10)
```

```python
# divide the data into x and y

X = df.iloc[:,:-1] # Independent column
y = df['Churn'] # Dependent column
```

```python
X
```

```python
y
```

```python
# splitting the data into train and test

from sklearn.model_selection import train_test_split
# stratify = y means to equally divide the values of the target column between # the training and testing
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 0 , stratify = y)
```

```python
'''
Train means TRAINING
Test means TESTING

x means QUESTIONS
y means Answers

x_train = TRAINING QUESTIONS  (Independent variables)
y_train=  TRAINING Answers    ( Dependent variable)

x_test=  TESTING QUESTIONS  (Independent variables)
y_test=  TESTING Answers   ( Dependent variable)
'''
```

```python
from random import Random
# Building the Random Forest Model

from sklearn.ensemble import RandomForestClassifier

classifier1 = RandomForestClassifier(n_estimators = 100)  # number of decision tree = 100


classifier1.fit(X_train,y_train)
```

```python
y_pred = classifier1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
df['Churn'].value_counts()
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

```python
'''            Actual
                 1      0

Predicted   1   TP     FP

            0   FN     TN


Accuracy = TP+TN / ( TP+TN+FP+FN)
Accuracy = 938 + 175 / 938 + 199 + 95 + 175
Accuracy = 0.7910447761194029
'''
```

```python
n = [100,200,300,400,500,600,700,800,900,1000]

accuracy = []

for i in n:
  model = RandomForestClassifier(n_estimators=i)
  model.fit(X_train,y_train)
  pred = model.predict(X_test)
  a = accuracy_score(y_test,pred) * 100
  accuracy.append(a)
```

```python
accuracy
```

```python
d = {"Number of trees" : n , "Accuracy" : accuracy}
df = pd.DataFrame(d)
df
```

# Random Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
 param_grid = {'n_estimators': [100,200,300],
               'max_depth' : [None,5,10],
             'min_samples_split':[2,5,10],
               'min_samples_leaf':[1,2,4],
              'criterion': ['gini','entropy']}
```

```python
rf = RandomForestClassifier(random_state = 0)
```

```python
# RandomizedsearchCv set up

random_search = RandomizedSearchCV(estimator = rf, param_distributions = param_grid, cv = 2, n_iter = 5)

# n_iter = 5 means it will  try 5 random combinations
```

```python
random_search.fit(X_train,y_train)
```

```python
random_search.best_params_
```

{'criterion': 'entropy', 'max_depth': 10, 'min_samples_leaf': 2, 'min_samples_split': 2, 'n_estimators': 200}

```python
best_model = random_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
